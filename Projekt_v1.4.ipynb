```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

import mandelbrot
```

### Fractal

```python
# Sampling area defined by its edges
area = [-2, 0.75, -1, 1]
mandel_set = mandelbrot.mandel(area)
```

```python
spacing = 0.005
max_iter = 100
```

```python
xx, yy = mandel_set.grid(spacing)
plane = mandel_set.generate_fractal(max_iter, spacing)
```

```python
fig = plt.figure()
ax = plt.gca()
pc = ax.pcolormesh(xx, yy, plane, cmap=plt.cm.get_cmap('RdBu'))
plt.show()
```

### Vary the number of iterations

```python
# Sampling area defined by its edges
area = [-2, 2, -2, 2]
mandel_set = mandelbrot.mandel(area)
```

#### Uniform sampling

```python
n_samples = 1000
iterations = np.logspace(1, 15, 15, base = 2)
```

```python
mandel_a_uniform = []
for i in iterations:
    np.random.seed(1)
    mandel_a_uniform.append(mandel_set.est_area(n_samples, int(i), 'uniform'))
```

```python
#Estimated area as a function of the number of iterations for uniform sampling
plt.figure()
plt.semilogx(iterations, np.array(mandel_a_uniform) - mandel_a_uniform[-1], base = 2, color = 'black')
plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Estimated Area', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.show()
```

#### Latin hypercube sampling

```python
mandel_a_lhs = []
for i in iterations:
    np.random.seed(1)
    mandel_a_lhs.append(mandel_set.est_area(n_samples, int(i), 'latin_hypercube'))
```

```python
# Estimated area as a function of the number of iterations for latin hypercube sampling
plt.figure()
plt.semilogx(iterations, np.array(mandel_a_lhs) - mandel_a_lhs[-1], base = 2, color = 'black')
plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Estimated Area', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.show()
```

#### Orthogonal sampling

#### Varying number of iterations for different sample sizes

```python
n_samples = [100, 1000, 10000]
```

```python
colors = ['black', 'red', 'blue']
plt.figure()

for s in range(len(n_samples)):
    mandel_a = []
    for i in iterations:
        np.random.seed(1)
        mandel_a.append(mandel_set.est_area(n_samples[s], int(i), 'uniform'))
    plt.semilogx(iterations, mandel_a, base = 2, color = colors[s], label = '%s samples'%n_samples[s])

plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Estimated Area', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend()
plt.show()
```
